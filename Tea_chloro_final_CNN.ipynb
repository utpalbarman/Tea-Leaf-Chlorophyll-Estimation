```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```python
print(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=42)
```

```python
print(X_train.shape)
```

```python
print(Y_test.shape)
```

```python
from keras.models import Model
from keras.layers import Dense, Input, Conv1D, Flatten, MaxPooling1D
#test_size = 0.40
# NN model
inputs = Input(shape=(12,1))
x = Conv1D(64, 2, padding='same', activation='elu')(inputs)
#x = MaxPooling1D(2)
x = Conv1D(128, 2, padding='same', activation='elu')(x)
x = Flatten()(x)
x = Dense(128, activation='elu')(x)
x = Dense(64, activation='elu')(x)
x = Dense(32, activation='elu')(x)
x = Dense(1, activation='linear')(x)
model = Model(inputs=[inputs], outputs=[x])
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mae'])
model.summary()
```

```python
model.fit(x=np.expand_dims(X_train, axis=2), y=Y_train, batch_size=10, epochs=10, verbose=1, shuffle=True)
```

```python
model.fit(x=np.expand_dims(X_test, axis=2), y=Y_test, batch_size=10, epochs=10, verbose=1, shuffle=True)

# model.fit(x=X_train, y=Y_train, batch_size=64, epochs=10, verbose=1, validation_split=0.1, shuffle=True)
# y_pred = model.predict(X_test)
```

```python
model.fit(x=np.expand_dims(X_train, axis=2), y=Y_train, batch_size=64, epochs=10, verbose=1, validation_split=0.2, shuffle=True)
```

```python
y_pred = model.predict(np.expand_dims(X_test, axis=2))
```

```python
dff = pd.DataFrame({'Actual': Y_test.flatten(), 'Predicted': y_pred.flatten()})
df1 = dff.head(20)
print(df1)
```

```python
print ('Mean Absolute Error(MAE)',metrics.mean_absolute_error(Y_test, y_pred))
print ('Mean Squared Error(MSE)',metrics.mean_squared_error(Y_test, y_pred))
print ('Root Mean Squared Error(RMSE)',np.sqrt(metrics.mean_squared_error(Y_test, y_pred)))     #test_size = 0.40
print('R Squared',metrics.r2_score(Y_test, y_pred))
plt.scatter(Y_test, y_pred)
plt.xlabel('True Values [chlorophyll]')
plt.ylabel('Predictions [chlorophyll]')
plt.axis('equal')
plt.axis('square')
plt.xlim([0,plt.xlim()[1]])
plt.ylim([0,plt.ylim()[1]])

_ = plt.plot([-100, 100], [-100, 100])
plt.savefig('D:/UTPAL PHD/Tea Leaf Journal Paper/Pic/1D_CNN.png', dpi=450, bbox_inches='tight')
```
